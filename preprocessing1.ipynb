```python
import pandas as pd
import os
```

```python
#각각의 내용을 담을 리스트 생성

def read_files(file_path, startline, endline) :
    lvList =[]
    vocaList = []
    classList =[]
    guideList = []

    with open(file_path, 'r', encoding='utf-8') as file :
        # 파일의 모든 줄을 읽어 오기 
        contents = file.readlines()
        # 189번 줄에 있는 내용부터 읽어와서 각각의 리스트에 넣기 위해 enumerate 사용
        for i, line in enumerate(contents[startline:endline], start=1):
            
            # 등급 담기    
            if i % 7 == 2 :
                #print(line)
                lvList.append(line.replace('\n', ''))
                
            # 어휘 담기
            elif i % 7 == 3 :
                #print(line)
                vocaList.append(line.replace('\n', ''))

            # 등급 담기
            elif i % 7 == 4 :
                #print(line)
                classList.append(line.replace('\n', ''))

            # 길잡이 말 담기       
            elif i % 7 == 5 :
                #print(line)
                guideList.append(line.replace('\n', ''))
            
            else : 
                print(f'{i}번째 줄은 공백입니다.')

        return lvList, vocaList, classList, guideList
```

```python
# txt 파일 경로 지정 ▶ 경로 확인
file_path1 = "./txt/1rate.txt"
#os.path.exists(file_path1)
```

```python
# 읽어올 줄과 마지막 줄+1 설정.
startline1 = 188
endline1 = 5332
lvList1,vocaList1,classList1,guideList1 = read_files(file_path1, startline1, endline1)

# print(len(lvList1), lvList1)
# print(len(vocaList1), vocaList1)
# print(len(classList1), classList1)
# print(len(guideList1), guideList1)
```

컬럼 : (등급 / 어휘 / 품사 / 길잡이 말)

```python
# 만들어놓은 등급, 어휘, 품사, 길잡이말 리스트를 하나의 데이터 프레임으로 생성.
def make_df(lvList, vocaList, classList, guideList) :
    col = ['등급', '어휘', '품사', '길잡이말']
    cnt = list(zip(lvList, vocaList, classList, guideList))
    return pd.DataFrame(cnt, columns=col)
```

```python
topik_1 = make_df(lvList1,vocaList1,classList1,guideList1)
```
